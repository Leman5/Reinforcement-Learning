```python
!pip install gym pyvirtualdisplay > /dev/null 2>&1
!apt-get install -y xvfb python-opengl ffmpeg > /dev/null 2>&1
!apt-get update > /dev/null 2>&1
!apt-get install cmake > /dev/null 2>&1
!pip install --upgrade setuptools 2>&1
!pip install ez_setup > /dev/null 2>&1
!pip install gym[atari] > /dev/null 2>&1

import gym
from gym.wrappers import Monitor
import glob
import io
import base64
from IPython.display import HTML
from pyvirtualdisplay import Display
from IPython import display as ipythondisplay

display = Display(visible=0, size=(1400, 900))
display.start()
```

```python
def show_video():
    mp4list = glob.glob('video/*.mp4')
    if len(mp4list) > 0:
        mp4 = mp4list[0]
        video = io.open(mp4, 'r+b').read()
        encoded = base64.b64encode(video)
        ipythondisplay.display(HTML(data='''<video alt="test" autoplay 
                loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii'))))
    else:
        print("Could not find video")


def wrap_env(env):
    env = Monitor(env, './video', force=True)
    return env
```

```python
env = wrap_env(gym.make("CartPole-v0"))

observation = env.reset()
score = 0 
episode = 1
while True:

    env.render()

    # your agent goes here
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    score+=reward
    if done:
        print('Episode:{} Score:{}'.format(episode, score))
        break

env.close()
show_video()
```

# Training RL model

```python
!mkdir Training
!mkdir Training/Logs
!mkdir Training/SavedModels
```

```python
log_path = os.path.join('Training','Logs')
```

```python
log_path
```

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda : env])
model = PPO('MlpPolicy',env,verbose=1,tensorboard_log=log_path)
#MlpPolicy = Multilevel perceptron policy
```

```python
model.learn(total_timesteps = 50000)
```

# Save and reload models

```python
ppo_path = os.path.join('Training','SavedModels','PPO_cartPole')
```

```python
#saving the model
model.save(ppo_path)
```

```python
#del current model and try to reload the saved model
del model
```

```python
model.learn(total_timesteps=2000)
```

```python
model = PPO.load(ppo_path,env = env)
```

# Evaluation

```python
evaluate_policy(model,env,n_eval_episodes=10,render=True)
#returns 2 parameters.... 
# 1. mean award  collected throught all the episodes
# 2. std deviation in the award collected
```

# Testing the model

```python
episodes = 5
for episode in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0 
    
    while not done:
        # env.render()
        action,_ = model.predict(obs) #instead of taking random actions we'll now take better actions acc to model observations
        state, reward, done, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
# env.close()
show_video()
```

# Training callback
use training callback to stop training the model when a particular reqard threshold value is met

```python
from stable_baselines3.common.callbacks import EvalCallback,StopTrainingOnRewardThreshold
```

```python
save_path = os.path.join('Training', 'Saved Models')
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200, verbose=1)
eval_callback = EvalCallback(env, 
                             callback_on_new_best=stop_callback, 
                             eval_freq=10000, 
                             best_model_save_path=save_path, 
                             verbose=1)
```

```python
model = PPO('MlpPolicy', env, verbose = 1, tensorboard_log=log_path)
model.learn(total_timesteps=20000, callback=eval_callback)
```

# Changing learning policies
▶ we can change the neural architechture

```python
net_arch=[dict(pi=[128, 128, 128, 128], vf=[128, 128, 128, 128])]
```

```python
model = PPO('MlpPolicy', env, verbose = 1, policy_kwargs={'net_arch': net_arch})
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```

```python
env.close()
```
